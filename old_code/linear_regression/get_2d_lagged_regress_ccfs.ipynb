```python
def get_2d_lr_plot():
    # Set up subplots figure with PlateCarree projection
    fig, ax = plt.subplots(1, 1, figsize=(4.5,3), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=180)}, dpi=100)
    
    vmin = -0.5
    vmax = 0.5
    
    lin_title = 'SST'

    # Plot Trends
    cbar1_data = SST_anom_lag.sel(lag=lag).plot(ax=ax,transform=ccrs.PlateCarree(),cmap=plt.get_cmap('RdBu').reversed(),
                           add_colorbar=False,vmin=vmin,vmax=vmax)
    
    # Give each plot gridlines, coastlines, and block out continents
    ax.gridlines()
    ax.coastlines()
#     ax.add_feature(cartopy.feature.LAND,zorder=100,facecolor='k',edgecolor='k')
    ax.set_aspect('auto')

    if lag == lag_list[0]:
        ax.set_title(lin_title, fontsize=20)
    elif lag == lag_list[-1]:
        cbar_ax1 = fig.add_axes([0.01, -0.1, 0.98, 0.05])
        cbar1 = fig.colorbar(cbar1_data, cax=cbar_ax1, orientation='horizontal', extend='both', ticks=[-0.5,0,0.5])
        cbar1.ax.tick_params(labelsize=20)
        cbar1.set_label(label='$K/K$', size=20)

        ax.set_title(None)
    else:
        ax.set_title(None)

    fig.suptitle('Lag = ' + str(lag), y=0.7, x=-0.01, rotation='vertical', fontsize=20);
```
