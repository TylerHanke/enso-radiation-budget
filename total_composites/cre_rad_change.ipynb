```python
# Load longwave and shortwave datasets for hiatus period
lw_flux = -CERES_flux.toa_lw_cldtyp_mon
lw_flux_clr = -CERES_flux.toa_lw_clr_mon

sw_flux = -CERES_flux.toa_sw_cldtyp_mon
sw_flux_clr = -CERES_flux.toa_sw_clr_mon

cld_area = CERES_cld_prop.cldarea_cldtyp_mon / 100

net_flux = lw_flux + sw_flux
net_flux_clr = lw_flux_clr + sw_flux_clr
```

```python
def get_mean_across_bins(x_cld,x_clr):
    arr = []
    
    for i in range(0,7,1):
        for j in range(0,6,1):            
            # Do calculation in each p,tau bin
            calc = cld_area.sel(press=i,opt=j) * (x_cld.sel(press=i,opt=j) - x_clr)
            
            # Append to arrays
            arr.append(calc)
    
    # Concat each array into DataArrays, and then sum
    calc_da = xr.concat(arr, dim='pressopt')
    calc_xr = calc_da.sum(dim='pressopt')
    
    return calc_xr
```

```python
lw_flux_calc = get_mean_across_bins(lw_flux,lw_flux_clr)
lw_diff = lw_flux_calc.mean(dim='time')

sw_flux_calc = get_mean_across_bins(sw_flux,sw_flux_clr)
sw_diff = sw_flux_calc.mean(dim='time')

net_flux_calc = get_mean_across_bins(net_flux,net_flux_clr)
net_diff = net_flux_calc.mean(dim='time')
```

```python
def get_ceres_cre_cld():
    # Set up subplots figure with PlateCarree projection
    fig, ax = plt.subplots(3,figsize=(18,12), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=180)})
    fig.tight_layout(h_pad=6)
    
    # Give each plot gridlines, coastlines, and block out continents
    for i in np.arange(0,3,1):
        ax[i].gridlines()
        ax[i].coastlines()
        #ax[i].add_feature(cartopy.feature.LAND,zorder=100,facecolor='k',edgecolor='k')

    # Plot Trends
    sw_diff.plot(ax=ax[0],transform=ccrs.PlateCarree(),cmap=plt.get_cmap('RdBu').reversed(),
                           cbar_kwargs={'shrink': 0.9},vmin=-100,vmax=100)
    ax[0].set_title('CERES SW UPWARD')
    
    lw_diff.plot(ax=ax[1],transform=ccrs.PlateCarree(),cmap=plt.get_cmap('RdBu').reversed(),
                           cbar_kwargs={'shrink': 0.9},vmin=-100,vmax=100)
    ax[1].set_title('CERES LW UPWARD')
    
    net_diff.plot(ax=ax[2],transform=ccrs.PlateCarree(),cmap=plt.get_cmap('RdBu').reversed(),
                           cbar_kwargs={'shrink': 0.9},vmin=-100,vmax=100)
    ax[2].set_title('CERES NET DOWNWARD')
```
